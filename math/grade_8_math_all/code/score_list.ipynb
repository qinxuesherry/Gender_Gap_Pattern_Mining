```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('../clean_data/after_clean.csv')
```

```python
data
```

```python
gap_stats = data.groupby('Jurisdiction')['Average scale score'].agg(['max', 'min']).reset_index()
gap_stats
```

```python
gap_stats['Min_Max'] = list(zip(gap_stats['min'], gap_stats['max']))

# Drop the separate 'max' and 'min' columns
gap_stats_combined = gap_stats.drop(columns=['max', 'min'])

gap_stats_combined
```

```python
gap_stats['MinMaxList'] = gap_stats['Min_Max'].apply(lambda x: list(x))
gap_stats_lists = gap_stats.drop(columns=['Min_Max', 'max', 'min'])
gap_stats_lists
```

```python
## gap in one column
score_list = data.groupby('Jurisdiction')['Average scale score'].apply(list).reset_index(name='Average scale score')

score_list
```

```python
score_list.to_csv('../clean_data/score_list.csv',index=False)
```
