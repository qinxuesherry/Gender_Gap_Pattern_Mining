```python
import pandas as pd

# Load the CSV file to examine its structure
score_list_df = pd.read_csv('../clean_data/score_list.csv')
score_list_df.head()
```

```python
# Correct the lambda function to handle float strings and convert them to integers
score_list_df['Average scale score'] = score_list_df['Average scale score'].apply(lambda x: [int(float(i)) for i in x.strip("[]").split(", ")])

# Define the years for which we want to expand the 'Average scale score' list into separate columns
years = [2003, 2005, 2007, 2009, 2011, 2013, 2015, 2017, 2019, 2022]

# Expand the 'Average scale score' list into separate columns for each year
for i, year in enumerate(years):
    score_list_df[str(year)] = score_list_df['Average scale score'].apply(lambda x: x[i] if i < len(x) else None)

# Drop the original 'Average scale score' column and the unnamed index column
score_list_df.drop(columns=['Average scale score'], inplace=True)

# Display the head of the modified DataFrame to verify the changes
score_list_df.head()
```

```python
score_list_df.to_csv('../clean_data/score_list_year.csv')
```
