```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('2000.csv')
data
```

```python
data.replace('—', np.nan, inplace=True)
data.replace('‡', np.nan, inplace=True)
```

```python
data['Score'] = pd.to_numeric(data['Score'], errors='coerce')
data
```

```python
df_2000=data[data['Year']=='2000']
df_2000_male=df_2000[df_2000['Gender']=='Male']
```

```python
df_2000_1=data[data['Year']=='2000_condition']
df_2000_1_male=df_2000_1[df_2000_1['Gender']=='Male']
```

```python
df_2000_male.plot(x='States', y='Score', marker='o', linestyle='-', color='blue')
df_2000_1_male.plot(x='States', y='Score', marker='o', linestyle='-', color='red')
```

```python
fig, ax = plt.subplots(figsize=(15, 13))
all_states = pd.concat([df_2000_male['States'], df_2000_1_male['States']]).unique()
min_score = min(df_2000_male['Score'].min(), df_2000_1_male['Score'].min())
max_score = max(df_2000_male['Score'].max(), df_2000_1_male['Score'].max())
y_ticks = np.arange(min_score, max_score + 1, 1)
# Plot for df_2000_male
df_2000_male.plot(x='States', y='Score', marker='o', linestyle='-', color='blue', ax=ax, label='2000 Male')

# Plot for df_2000_1_male on the same axes
df_2000_1_male.plot(x='States', y='Score', marker='o', linestyle='-', color='red', ax=ax, label='2000 Condition Male')

plt.title('Male Scores in 2000 by State for Different Conditions')
plt.xlabel('State')
plt.ylabel('Score')
ax.set_xticks(range(len(all_states)))  # Setting x-ticks positions
ax.set_xticklabels(all_states, rotation=90) 
ax.set_yticks(y_ticks)
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_2000_female=df_2000[df_2000['Gender']=='Female']
df_2000_1_female=df_2000_1[df_2000_1['Gender']=='Female']
```

```python
fig, ax = plt.subplots(figsize=(15, 13))
all_states1 = pd.concat([df_2000_female['States'], df_2000_1_female['States']]).unique()
min_score1 = min(df_2000_female['Score'].min(), df_2000_1_female['Score'].min())
max_score1 = max(df_2000_female['Score'].max(), df_2000_1_female['Score'].max())
y_ticks = np.arange(min_score1, max_score1 + 1, 1)
# Plot for df_2000_male
df_2000_female.plot(x='States', y='Score', marker='o', linestyle='-', color='blue', ax=ax, label='2000 Female')

# Plot for df_2000_1_male on the same axes
df_2000_1_female.plot(x='States', y='Score', marker='o', linestyle='-', color='red', ax=ax, label='2000 Condition Female')

plt.title('Female Scores in 2000 by State for Different Conditions')
plt.xlabel('State')
plt.ylabel('Score')
ax.set_xticks(range(len(all_states1)))  # Setting x-ticks positions
ax.set_xticklabels(all_states1, rotation=90) 
ax.set_yticks(y_ticks)
plt.legend()
plt.grid(True)
plt.show()
```
