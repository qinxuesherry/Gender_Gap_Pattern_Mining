```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('after_clean.csv')
data.head()
```

```python
male_scores = data[data['Gender'] == 'Male'].rename(columns={'Average scale score': 'Male Score'}).drop('Gender', axis=1)
female_scores = data[data['Gender'] == 'Female'].rename(columns={'Average scale score': 'Female Score'}).drop('Gender', axis=1)

# Merge the datasets on 'Year' and 'Jurisdiction'
merged_data = pd.merge(male_scores, female_scores, on=['Year', 'Jurisdiction'])
merged_data['Gap'] = merged_data['Male Score'] - merged_data['Female Score']
merged_data.head()
```

```python
merged_data.to_csv('merged_data.csv')
```
