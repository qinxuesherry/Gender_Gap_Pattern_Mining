```python
import pandas as pd

# Load the CSV file to examine its structure
gap_list_df = pd.read_csv('../clean_data/gap_list.csv')
gap_list_df.head()
```

```python
# Correct the lambda function to handle float strings and convert them to integers
gap_list_df['Gaps'] = gap_list_df['Gaps'].apply(lambda x: [int(float(i)) for i in x.strip("[]").split(", ")])

# Define the years for which we want to expand the 'Gaps' list into separate columns
years = [2000, 2003, 2005, 2007, 2009, 2011, 2013, 2015, 2017, 2019, 2022]

# Expand the 'Gaps' list into separate columns for each year
for i, year in enumerate(years):
    gap_list_df[str(year)] = gap_list_df['Gaps'].apply(lambda x: x[i] if i < len(x) else None)

# Drop the original 'Gaps' column and the unnamed index column
gap_list_df.drop(columns=['Unnamed: 0', 'Gaps'], inplace=True)

# Display the head of the modified DataFrame to verify the changes
gap_list_df.head()
```

```python
gap_list_df.to_csv('../clean_data/gap_list_year.csv')
```
