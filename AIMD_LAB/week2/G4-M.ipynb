```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('G4-M-Gen.csv')
df.head()
```

```python
pivoted_data = df.pivot_table(index=['Year', 'Jurisdiction'], columns='Gender', values='Average scale score')
pivoted_data.reset_index(inplace=True)
pivoted_data
pivoted_data.to_csv('pivoted_data.csv')
```

```python
plt.figure()
plt.plot(pivoted_data['Year'], pivoted_data['Male'], label='Male', marker='o')
plt.plot(pivoted_data['Year'], pivoted_data['Female'], label='Female', marker='o')
plt.style.use('seaborn-darkgrid')

plt.xlabel("Year")
plt.ylabel("Average Score")
plt.title("Average Mathematics Score By Gender")
plt.xticks(pivoted_data['Year'], rotation=45) 
plt.legend(title='Gender')
plt.tight_layout()

plt.show()
```
