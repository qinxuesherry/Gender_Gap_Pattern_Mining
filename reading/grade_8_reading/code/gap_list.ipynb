```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('../clean_data/gap.csv')
data.head()
```

```python
gap_stats = data.groupby('Jurisdiction')['Gap'].agg(['max', 'min']).reset_index()
gap_stats
```

```python
gap_stats['Min_Max'] = list(zip(gap_stats['min'], gap_stats['max']))

# Drop the separate 'max' and 'min' columns
gap_stats_combined = gap_stats.drop(columns=['max', 'min'])

gap_stats_combined
```

```python
gap_stats['MinMaxList'] = gap_stats['Min_Max'].apply(lambda x: list(x))
gap_stats_lists = gap_stats.drop(columns=['Min_Max', 'max', 'min'])
gap_stats_lists
```

```python
## gap in one column
gap_list = data.groupby('Jurisdiction')['Gap'].apply(list).reset_index(name='Gaps')

gap_list
```

```python
gap_list.to_csv('../clean_data/gap_list.csv', index=False)
```
